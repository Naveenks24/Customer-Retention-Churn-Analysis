```python
import pandas as pd

# Load the file into memory (matching your exact file name in the sidebar)
df = pd.read_csv('Telco-Customer-Churn.csv')

# Show the first 5 rows to confirm it worked
display(df.head())
```

```python
# 1. Force 'TotalCharges' to be a number. 
# errors='coerce' tells Python: "If you find a weird blank space, turn it into a standard NaN (Not a Number/Null) instead of crashing."
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# 2. Now let's check how many missing (NaN) values we have in every column
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
# 1. Drop the 11 rows with missing TotalCharges
df.dropna(inplace=True)

# 2. Let's calculate the overall Churn rate
churn_counts = df['Churn'].value_counts()
churn_percentage = df['Churn'].value_counts(normalize=True) * 100

print("--- Overall Customer Churn ---")
print(churn_counts)
print("\nPercentage:\n", churn_percentage)
```

```python
# Create a cross-tabulation (crosstab) to see the churn breakdown per contract type
# normalize='index' calculates the percentage WITHIN each contract row
contract_churn = pd.crosstab(df['Contract'], df['Churn'], normalize='index') * 100

print("--- Churn Percentage by Contract Type ---")
# .round(2) just makes the output cleaner by limiting to 2 decimal places
display(contract_churn.round(2))
```

```python
import matplotlib.pyplot as plt

# We will plot just the 'Yes' column (the actual churn rate)
churn_rates = contract_churn['Yes']

# Set up the canvas size
plt.figure(figsize=(8, 5))

# Create the bar chart with some custom colors
bars = churn_rates.plot(kind='bar', color=['#ff9999', '#66b3ff', '#99ff99'], edgecolor='black')

# Add titles and labels
plt.title('Customer Churn Rate by Contract Type', fontsize=14, fontweight='bold')
plt.ylabel('Churn Percentage (%)', fontsize=12)
plt.xlabel('Contract Type', fontsize=12)

# Keep the x-axis labels flat for easy reading
plt.xticks(rotation=0) 

# This clever loop adds the exact data labels on top of each bar!
# This corrected loop uses f-string formatting {:.1f} to round to 1 decimal place
for index, value in enumerate(churn_rates):
    plt.text(index, value + 1, f"{value:.1f}%", ha='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the canvas
plt.figure(figsize=(8, 6))

# Create the boxplot
# x is our category (Churn Yes/No), y is the number we are analyzing (MonthlyCharges)
sns.boxplot(x='Churn', y='MonthlyCharges', data=df, palette=['#66b3ff', '#ff9999'])

# Add titles and labels
plt.title('Impact of Monthly Charges on Customer Churn', fontsize=14, fontweight='bold')
plt.xlabel('Did the Customer Churn?', fontsize=12)
plt.ylabel('Monthly Charges ($)', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the canvas
plt.figure(figsize=(10, 6))

# Create a histogram showing tenure, split by whether they churned or not
sns.histplot(data=df, x='tenure', hue='Churn', multiple='layer', 
             bins=36, palette=['#66b3ff', '#ff9999'], kde=True)

# Add titles and labels
plt.title('When Do Customers Leave? (Customer Tenure vs. Churn)', fontsize=14, fontweight='bold')
plt.xlabel('Tenure (Months with the Company)', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```
